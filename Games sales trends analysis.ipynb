# Описание проекта

Интернет-магазин X продаёт компьютерные игры (весь мир).
Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы.

ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр в США и Канаде.

# Цель проекта

Нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

Предоставлены данные до 2016 года. Нужно отработать принцип работы с данными при прогнозировании продаж на 2017 год по данным 2016-го.

### 2.3. Обработка пропусков.

```python
#выведем долю игр и продаж с отсуствующими значениями в 'name','year_of_release','genre','critic_score','user_score','rating'
for col in ['name','year_of_release','genre','critic_score','user_score','rating']:
    print(f"Игры с NAN в {col}. \nКоличество = {df[col].isna().sum()}\n% от количества значений датасета = {(df[col].isna().sum()/len(df[col])):.2%}")
    print(f"Доля от продаж na_sales:{df[df[col].isna()]['na_sales'].sum()/df['na_sales'].sum():.2%}")
    print(f"Доля от продаж eu_sales:{df[df[col].isna()]['eu_sales'].sum()/df['eu_sales'].sum():.2%}")
    print(f"Доля от продаж jp_sales:{df[df[col].isna()]['jp_sales'].sum()/df['jp_sales'].sum():.2%}")
    print(f"Доля от продаж other_sales {df[df[col].isna()]['other_sales'].sum()/df['other_sales'].sum():.2%}")
    print()
```

```python
# Удалим строки с пропусками в данных 'name','year_of_release','genre'
df_d = df.copy()
for col in ['name','year_of_release','genre']:
    df_d = df_d[df_d[col].notna()]
```

Пропусков в `name` и `genre` немного, удалили строки, это некритично для дальнейшего анализа.

Пропусков в `year_of_release` - 268, это 1.6% от строк в датасете, при этом год критичен для анализа и прогноза в разрезе лет. Поэтому удалили строки с пропусками годов выпуска.

#### Обработка значений `tbd`

```python
# По сути `tbd` это to be determined, i.e. значение NA, поэтому заменим его на NAN
df_d['user_score'].replace('tbd',np.NaN,inplace = True)

print('Поле :user_score\nуникальных значений:',len(df_d['user_score'].unique()), '\nNaN: ', df_d['user_score'].isna().sum(),'\n') 
print(df_d.groupby(['user_score'])['user_score'].count().sort_values(ascending = False))
print(df_d[df_d['user_score']=='tbd'])
```

```python
# Выведем строки с пропусками в данных 'critic_score','user_score','rating'
df_na = df_d.copy()
df_na = df_d[(df_d['critic_score'].isna())&(df_d['user_score'].isna())&(df_d['rating'].isna())]
df_na
```

```python
sns.set(style="whitegrid")
for x in ['platform','year_of_release','genre']:
    a = sns.histplot(df[df[x].notna()][x],binwidth = 1)
    b = sns.histplot(df_na[x],color = 'red',binwidth = 1)
    plt.xticks(rotation=90)
    plt.xlabel(f'Красным - количество NaN среди {x}')
    plt.show()
```

```python
df_after_2000 = df[df['year_of_release'] > 2000]
df_before_2000 = df[df['year_of_release'] <= 2000]

df_na_after_2000 = df_na[df_na['year_of_release'] > 2000]
df_na_before_2000 = df_na[df_na['year_of_release'] <= 2000]

for x in ['platform','genre']:
    plt.title(f'Распределение отсутствия значений {x} до 2000г', size = 16)
    sns.histplot(df_before_2000[df_before_2000[x].notna()], x = df_before_2000[df_before_2000[x].notna()][x] ,multiple="stack", binwidth = 1, )
    sns.histplot(df_na_before_2000, x = df_na_before_2000[x] ,multiple="stack", binwidth = 1, color = 'orange')
    plt.xticks(rotation=90)
    plt.show()

for x in ['platform','genre']:
    plt.title(f'Распределение отсутствия значений {x} после 2000г', size = 16)
    sns.histplot(df_after_2000[df_after_2000[x].notna()], x = df_after_2000[df_after_2000[x].notna()][x] ,multiple="stack", binwidth = 1, )
    sns.histplot(df_na_after_2000, x = df_na_after_2000[x] ,multiple="stack", binwidth = 1, color = 'orange')
    plt.xticks(rotation=90)
    plt.show()
```

#### Причины, приведшие к пропускам;
После 2000г возрастает количество игр и возникает бОльшая часть пропусков в оценках критиков, пользователей и рейтингах.
Возможно, причиной тому резкий рост количества выходящих игр.

### 5.1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые?

H0 - средние пользовательские рейтинги платформ Xbox One и PC равны

H1 - средние пользовательские рейтинги платформ Xbox One и PC НЕ равны

```python
alpha = .05 # критический уровень статистической значимости

usr_rating_xbox = df_ex[df_ex['platform'] == 'XOne']['user_score'].dropna()
usr_rating_pc = df_ex[df_ex['platform'] == 'PC']['user_score'].dropna()

# проверим выборки на нормальность
results_xbox = st.shapiro(usr_rating_xbox)
results_pc = st.shapiro(usr_rating_pc)

if (results_xbox[1] < alpha):
    print(f"Значение p-value = {results_xbox[1]:.6f}. Отвергаем нулевую гипотезу: распределение XBOX не нормально")
else:
        print(f"Значение p-value = {results_xbox[1]:.6f}. Не получилось отвергнуть нулевую гипотезу, распределение XBOX нормально")

if (results_pc[1] < alpha):
    print(f"Значение p-value = {results_pc[1]:.6f}. Отвергаем нулевую гипотезу: распределение PC не нормально")
else:
        print(f"Значение p-value = {results_pc[1]:.6f}. Не получилось отвергнуть нулевую гипотезу, распределение PC нормально")

plt.figure(figsize = (10,4))
sns.histplot(usr_rating_xbox,
            ax = plt.subplot(1,2,1))
plt.title('usr_rating_xbox',size = 15)

sns.histplot(usr_rating_pc,
            ax = plt.subplot(1,2,2))
plt.title('usr_rating_pc',size = 15)
plt.show()
```

</font><font color='Blue'><b>Тест Стьюдента</b></font><br>

```python
# Расчёт с помошью критерия Стьюдента
alpha = .05 # критический уровень статистической значимости
print(f'usr_rating_xbox количество значений: {len(usr_rating_xbox)}\n дисперсия:{np.var(usr_rating_xbox)}\n')
print(f'usr_rating_pc количество значений: {len(usr_rating_pc)}\n дисперсия:{np.var(usr_rating_pc)}\n')


results = st.ttest_ind(usr_rating_xbox, usr_rating_pc)

if results.pvalue < alpha:
    print(f'p-value = {results[1]:.8f}\nОтвергаем нулевую гипотезу о равенстве средних пользовательских рейтингов платформ Xbox One и PC')
else:
    print(
        f'p-value = {results[1]:.8f}\nНе получилось отвергнуть нулевую гипотезу о равенстве средних пользовательских рейтингов платформ Xbox One и PC'
    )
```
